```python
def box_plot(df):
    
    df_obj = df.select_dtypes(include='object')
    df_ints = df.drop(df_obj.columns,axis = 1)
    
    n = int(np.ceil(df_ints.shape[1]/2))
    
    fig, axes = plt.subplots(n,2, figsize = (16,6*n))
    fig.suptitle('Useful Boxplots', fontsize = 18)
    c = 0
    for i in range(0,n):
        sns.boxplot(data = df_ints, x = df_ints.columns[c], ax = axes[i,0])
        sns.boxplot(data = df_ints, x = df_ints.columns[c+1], ax = axes[i,1])
        c = c + 2
    plt.show()
```

```python
box_plot(df)
```
